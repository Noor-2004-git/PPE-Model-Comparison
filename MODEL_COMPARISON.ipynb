```python
from ultralytics import YOLO
import os
import glob
import pandas as pd
import numpy as np
from collections import defaultdict
```

```python
MODELS = {
    "Model1": "/content/models/model1.pt",
    "Model2": "/content/models/model2.pt",
    "Model3": "/content/models/model3.pt"
}

IMG_DIR = "/content/test_dataset/images"
LBL_DIR = "/content/test_dataset/labels"

SAVE_DIR = "/content/drive/MyDrive/PPE_Model_Comparison"

os.makedirs(SAVE_DIR, exist_ok=True)
```

```python
# Dataset classes
GT_MAP = {
    1: "gloves",
    3: "helmet",
    9: "vest"
}

MODEL_MAP = {
    "Model1": {
        "Gloves": "gloves",
        "helmet": "helmet",
        "Vest": "vest"
    },

    "Model2": {
        "gloves": "gloves",
        "hardhat": "helmet",
        "safety vest": "vest"
    },

    "Model3": {
        "Gloves": "gloves",
        "Hardhat": "helmet",
        "Safety Vest": "vest"
    }
}
```

```python
def load_gt(label_file):

    objs = []

    with open(label_file) as f:
        lines = f.readlines()

    for line in lines:

        parts = line.strip().split()

        cls = int(parts[0])

        if cls in GT_MAP:
            objs.append(GT_MAP[cls])

    return objs
```

```python
results = []

for model_name, model_path in MODELS.items():

    print("\nEvaluating", model_name)

    model = YOLO(model_path)

    TP = defaultdict(int)
    FP = defaultdict(int)
    FN = defaultdict(int)

    image_paths = sorted(glob.glob(f"{IMG_DIR}/*.jpg"))

    for img_path in image_paths:

        label_path = os.path.join(
            LBL_DIR,
            os.path.basename(img_path).replace(".jpg",".txt")
        )

        gt = load_gt(label_path)

        pred = []

        res = model.predict(
            img_path,
            conf=0.25,
            verbose=False
        )

        for c in res[0].boxes.cls.cpu().numpy():

            cls_name = model.names[int(c)]

            if cls_name in MODEL_MAP[model_name]:
                pred.append(
                    MODEL_MAP[model_name][cls_name]
                )

        gt_set = set(gt)
        pred_set = set(pred)

        for cls in ["helmet","gloves","vest"]:

            if cls in gt_set and cls in pred_set:
                TP[cls] += 1

            elif cls not in gt_set and cls in pred_set:
                FP[cls] += 1

            elif cls in gt_set and cls not in pred_set:
                FN[cls] += 1

    rows = []

    macro_f1 = []

    for cls in ["helmet","gloves","vest"]:

        tp = TP[cls]
        fp = FP[cls]
        fn = FN[cls]

        precision = tp/(tp+fp+1e-9)
        recall    = tp/(tp+fn+1e-9)

        f1 = 2*precision*recall/(precision+recall+1e-9)

        macro_f1.append(f1)

        rows.append([
            model_name,
            cls,
            tp,
            fp,
            fn,
            precision,
            recall,
            f1
        ])

    df = pd.DataFrame(
        rows,
        columns=[
            "Model",
            "Class",
            "TP",
            "FP",
            "FN",
            "Precision",
            "Recall",
            "F1"
        ]
    )

    df.to_csv(
        f"{SAVE_DIR}/{model_name}_metrics.csv",
        index=False
    )

    results.append({
        "Model": model_name,
        "Macro_F1": np.mean(macro_f1)
    })
```

```python
ranking = pd.DataFrame(results)

ranking = ranking.sort_values(
    by="Macro_F1",
    ascending=False
)

ranking
```

```python
ranking.to_csv(
    f"{SAVE_DIR}/final_ranking.csv",
    index=False
)

ranking
```
